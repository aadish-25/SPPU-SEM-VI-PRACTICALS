```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
print("Dataset loaded...")
```

```python
# Dimenstion or shape of the dataset
print("Dataset shape: ", df.shape)

# Columns
print("Columns in the dataset: ", df.columns.tolist())
```

```python
# Structural Summary
df.info()
```

```python
# Statistical Summary
df.describe()
```

```python
# Datatypes of variables
print("Data types of the variables: ")
print(df.dtypes)
```

```python
# Missing values
missing_values = df.isnull().sum()
print("Missing values per column: \n")
print(missing_values)
```

```python
# Drop missing values (not recommended as the dataset would reduce to a very small size)

df.dropna()

# We avoid dropping the entire column as it wud shorten the dataset
if "Cabin" in df.columns:
    df = df.drop(columns=["Cabin"])
    print("Cabin dropped")
else:
    print("Cabin already dropped")
    

# Fills the null values with median of the Age column, inplace makes perm changes to the df, which not recommended
df["Age"] = df["Age"].fillna(df["Age"].median()) 

# Replaces the null values with the mode of the Embarked col
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0]) 

df.isnull().sum()
```

```python
df.head(5)
```

```python
# Converting approprite columns into categories

df["Sex"] = df["Sex"].astype("category")
df["Embarked"] = df["Embarked"].astype("category")
df["Pclass"] = df["Pclass"].astype("category")
df["Survived"] = df["Survived"].astype("category")

df.dtypes
```

```python
# Converts categorical columns into numerical binary columns using One-Hot Encoding.

# df = pd.get_dummies(df, ...) -> alters the current df itself
df = pd.get_dummies(
    df,
    columns=["Sex", "Embarked", "Pclass"],
)

df.head(2)
```

```python
# Normalization

scaler = MinMaxScaler()

# Normalizes Age and Fare columns to values between 0 and 1
df[["Age", "Fare"]] = scaler.fit_transform(
    df[["Age", "Fare"]]
)
```
