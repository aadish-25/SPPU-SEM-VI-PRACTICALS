```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
iris = sns.load_dataset("iris")
print(iris.head())
```

```python
iris.hist(figsize=(10,8), bins=20)

plt.suptitle("Histogram of Iris Dataset Features")

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris)
plt.title("Boxplot of Iris Dataset Features")

plt.show()
```

```python
numeric_cols = iris.select_dtypes(include="number").columns

for col in numeric_cols:
    Q1 = iris[col].quantile(0.25)
    Q3 = iris[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers = iris[(iris[col] < lower_limit) | (iris[col] > upper_limit)]

    print(f"\nFeature: {col}")
    print(f"--------------------")
    print(f"Q1: {Q1}")
    print(f"Median: {iris[col].median()}")
    print(f"Q3: {Q3}")
    print(f"IQR: {IQR}")
    print(f"Lower Limit: {lower_limit:.2f}")
    print(f"Upper Limit: {upper_limit:.2f}")
    print(f"Number of Outliers: {len(outliers)}")


    if len(outliers) > 0:
        print("Outlier values:")
        print(outliers[[col, "species"]])
```
