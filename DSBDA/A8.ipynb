```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(3)
```

```python
titanic.info
```

```python
sns.histplot(titanic['fare'], bins=30, kde=True)
plt.title('Distribution of Titanic Ticket Fare')    
plt.xlabel("Fare")
plt.ylabel("No of Passenger")
```

```python
sns.countplot(x="sex", data=titanic)
plt.title("Number of passangers acc to gender")
```

```python
sns.countplot(x="sex", hue="survived", data=titanic)
# 0 = not survived, 1 = survived
plt.title("Survival acc to gender")
plt.show()
```

```python
sns.countplot(x="class", hue="survived", data=titanic)

plt.title("Passenger Class vs Survival")
plt.show()
```

```python
sns.histplot(titanic["age"], bins=20, kde=True)

plt.title("Age Distribution")
plt.show()
```

```python
corr = titanic.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
