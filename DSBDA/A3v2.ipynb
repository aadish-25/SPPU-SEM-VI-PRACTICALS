```python
import pandas as pd
import seaborn as sns
```

```python
data = {
    "name": ["Amit", "Rahul", "Sneha", "Priya", "Rohan","Neha", "Akash", "Meera", "Karan", "Pooja"],
    "attendance": [92, 85, 95, 88, 76,91, 84, 89, 73, 96],
    "gender": ["Male", "Male", "Female", "Female", "Male","Female", "Male", "Female", "Male", "Female"]
}
```

```python
df = pd.DataFrame(data)
```

```python
grouped = df.groupby("gender")["attendance"]
```

```python
summary = grouped.agg(['mean', 'median', 'std', 'min', 'max'])
```

```python
summary
```

```python
male = df[df["gender"]=="Male"]["attendance"].tolist()
female = df[df["gender"]=="Female"]["attendance"].tolist()
```

```python
print("\nMale Attendance List:")
print(male)

print("\nFemale Attendance List:")
print(female)
```

```python
iris = sns.load_dataset('iris')
```

```python
species = set(iris['species'])
```

```python
iris_group = iris.groupby("species")
```

```python
summary = iris_group.agg([
    'min', 'max', 'mean', 'median', 'std'
])
summary['petal_length']
```

```python
for col in iris.select_dtypes(include="number").columns:
    print(f"\n==========Statistics of {col}==========")
    print(summary[col])
```

```python
for species_name, data in iris_group:
    print(species_name)
    print()
    print(data.describe())
    print("=============")
```
