```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
df = pd.read_csv('adult-census-data.csv')
df
df.count()
```

```python
print(df.isnull().sum())
```

```python
# This tells you how many rows in each column have a "?"
print((df == "?").sum())

# NOw we replace all the ? by NaN and then drop those rows
df = df.replace('?', np.nan)
df = df.dropna()
```

```python
print("COlumns list :" ,df.columns.tolist())
df.info()
```

#### GPT

```python
from tabulate import tabulate

# 1. Calculate - Group by both Marital Status AND Sex
summary = df.groupby(['marital.status', 'sex'])['age'].agg(['mean', 'median', 'min', 'max', 'std']).reset_index()

# 2. Rename columns for clarity
summary.columns = ['Marital Status', 'Sex', 'Avg', 'Med', 'Min', 'Max', 'Std']

# 3. Format and Print
print("-" * 70)
print("CENSUS DATA: AGE STATISTICS BY MARITAL STATUS AND GENDER")
print("-" * 70)

print(tabulate(summary.round(2), headers='keys', tablefmt='psql', showindex=False))
```
